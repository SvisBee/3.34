```python
"""ru
Описание
Эта программа предназначена для получения и отображения свойств химических элементов
с использованием библиотеки Mendeleev для получения данных об элементах и библиотеки
Matplotlib для визуализации. Пользователь вводит символ элемента, и программа выводит
его свойства и визуализирует цвета в различных цветовых схемах.

Импортируемые библиотеки
Matplotlib
matplotlib.pyplot: используется для создания графиков.
matplotlib.patches: используется для добавления фигур (например, прямоугольников) на график.
Mendeleev
mendeleev.element: функция, которая возвращает объект элемента, содержащий его свойства.
"""



"""en
Description
This program is designed to obtain and display the properties of chemical elements
using the Mendeleev library for obtaining element data and the Matplotlib library
for visualization. The user enters the symbol of an element, and the program outputs
its properties and visualizes colors in various color schemes.

Imported libraries
Matplotlib
matplotlib.pyplot: Used to create graphs.
matplotlib.patches: Used to add shapes (e.g. rectangles) to a graph.
Mendeleev
mendeleev.element: A function that returns an element object containing its properties.
"""

import matplotlib.pyplot as plt
import matplotlib.patches as patches
from mendeleev import element

class ElementProperties:

    """ ru
    Класс для представления свойств и характеристик химического элемента.

    Атрибуты:
    ----------
    symbol : str
        Символ химического элемента (например, 'H' для водорода).

    Методы:
    -------
    get_properties():
        Возвращает словарь со свойствами и характеристиками элемента.
    """



    """ en
    A class for representing the properties and characteristics of a chemical element.

    Attributes:
    ----------
    symbol : str
      is a symbol of a chemical element (for example, 'H' for hydrogen).

    Methods:
    -------
    get_properties():
      Returns a dictionary with the properties and characteristics of the element.
    """

    def __init__(self, symbol):

        """ru
        Инициализирует объект ElementProperties с заданным символом элемента.

        Параметры:
        ----------
        symbol : str
            Символ химического элемента.
        """



        """en
        Initializes the Element Properties object with the specified element symbol.

        Parameters:
        ----------
        symbol : str
          is a symbol of a chemical element.
        """

        self.symbol = symbol
        self.element = element(symbol)

    def get_properties(self):

        """ru
        Получает свойства и характеристики элемента.

        Возвращает:
        ----------
        dict
            Словарь со свойствами и характеристиками элемента.
        """



        """en
        Retrieves the properties and characteristics of the element.

        Returns:
        ----------
        dict
            A dictionary with the properties and characteristics of the element.
        """

        properties = {
            'Name': getattr(self.element, 'name', None),
            'Name original': getattr(self.element, 'name_origin', None),
            'Family': self.determine_family(),
            'Discoverers': getattr(self.element, 'discoverers', None),
            'Discovery location': getattr(self.element, 'discovery_location', None),
            'Discovery year': getattr(self.element, 'discovery_year', None),
            'Atomic number': getattr(self.element, 'atomic_number', None),
            'Atomic weight': getattr(self.element, 'atomic_weight', None),
            'Group': getattr(self.element, 'group_id', None),
            'Period': getattr(self.element, 'period', None),
            'Block': getattr(self.element, 'block', None),
            'Density': getattr(self.element, 'density', None),
            'Melting point': getattr(self.element, 'melting_point', None),
            'Boiling point': getattr(self.element, 'boiling_point', None),
            'Electronegativity Pauling': getattr(self.element, 'en_pauling', None),
            'Atomic radius': getattr(self.element, 'atomic_radius', None),
            'Covalent radius': getattr(self.element, 'covalent_radius', None),
            'Van der Waals radius': getattr(self.element, 'vdw_radius', None),
            'specific heat capacity': getattr(self.element, 'specific_heat_capacity', None),
            'Thermal conductivity': getattr(self.element, 'thermal_conductivity', None),
            'Electronic configuration': getattr(self.element, 'ec', None),
            'Color (CPK)': getattr(self.element, 'cpk_color', None),
            'Color (Jmol)': getattr(self.element, 'jmol_color', None),
            'Color (MOLCAS GV)': getattr(self.element, 'molcas_gv_color', None),
        }
        return properties

    def determine_family(self):

            """ru
            Определяет семейство элемента (металл, неметалл и т.д.).

            Возвращает:
            ----------
            str
                Семейство элемента.
            """


            """en
            Defines the family of the element (metal, nonmetal, etc.).

            Returns:
            ----------
            str
              is the family of the element.
            """

            if self.element.atomic_number in {1, 6, 7, 8, 15, 16, 34} or self.element.period in {17, 18}:
              return "Nonmetal"
            elif self.element.atomic_number in {5, 14, 32, 33, 51, 52, 84}:
              return "Semi-metal"
            elif self.element.atomic_number in range(57, 72):
              return "Lanthanide"
            elif self.element.atomic_number in range(89, 104):
              return "Actinoid"
            else:
              return "Metal"

    def display_colors(self):

      """ru
      Отображает цвета элемента в выбранных цветовых схемах.
      """



      """en
      Displays the colors of the element in the selected color schemes.
      """

      colors = {
          'CPK': getattr(self.element, 'cpk_color', None),
          'Jmol': getattr(self.element, 'jmol_color', None),
          'MOLCAS GV': getattr(self.element, 'molcas_gv_color', None),
      }

      fig, ax = plt.subplots()
      for i, (name, color) in enumerate(colors.items()):
          if color:
              ax.add_patch(patches.Rectangle((i * 2, 0), 2, 2, color=color, label=name))

      ax.set_xlim(0, 6)
      ax.set_ylim(0, 2)
      ax.set_aspect('equal')
      plt.legend(loc='upper left')
      plt.show(block=False)
      plt.pause(1)  # Wait for 1 second to ensure the window appears
      plt.close(fig) # Close the figure to allow further inputs

      print('\n\n\n')


while True:
    symbol = input("Enter the symbol of an element (or 'exit' to quit): ")
    if symbol.lower() == 'exit':
        break
    try:
      element_props = ElementProperties(symbol)
      properties = element_props.get_properties()

      # Output of properties
      for prop, value in properties.items():
        print(f"{prop}: {value}")

      # Displaying colors
      element_props.display_colors()
    except:
      print(f"Element with symbol '{symbol}' not found. Please enter a valid element symbol.")
```
